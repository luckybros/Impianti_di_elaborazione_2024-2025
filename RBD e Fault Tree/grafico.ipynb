```python
import numpy as np
import matplotlib.pyplot as plt

MTTF = 800  
lambda_val = 1 / MTTF  
t = np.linspace(0, 2000, 200)  
R = np.exp(-lambda_val * t) 

R_sys1 = 1 - (1 - R**4)**2
R_sys2 = (1 - (1 - R)**2)**4

plt.figure(figsize=(10, 6))
plt.plot(t, R_sys1, label=r'$R_{sys1}$', linestyle='-', linewidth=2)
plt.plot(t, R_sys2, label=r'$R_{sys2}$', linestyle='-', linewidth=2)

plt.xlabel('Tempo (ore)')
plt.ylabel('Affidabilit√† del sistema')
plt.title('Confronto tra $R_{sys1}$ e $R_{sys2}$ con MTTF = 800 ore')
plt.legend()
plt.grid()

plt.show()
```

```python
t = np.linspace(0, 30000, 500)

R_sys = np.exp(-t/6000) * np.exp(-t/7000) * (1 - (1 - np.exp(-t/10000)) * (1 - np.exp(-t/5000)))
R_simplex = np.exp(-t/6000) * np.exp(-t/7000) * np.exp(-t/10000)

plt.figure(figsize=(10, 6))
plt.plot(t, R_sys, label=r'$R_{sys}$', color='b', linewidth=2)
plt.plot(t, R_simplex, label=r'$R_{simplex}$', color='r', linestyle='-', linewidth=2)

plt.title('Comparison of $R_{sys}$ and $R_{simplex}$', fontsize=14)
plt.xlabel('Time (t)', fontsize=12)
plt.ylabel('R(t)', fontsize=12)
plt.legend()
plt.grid(True)

plt.show()
```
